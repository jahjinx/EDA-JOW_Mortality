## The Data
The data used in this project was obtained from [Mendeley Data](https://data.mendeley.com/datasets/xwj98nb39r/1), with additional information available on [ScienceDirect](https://www.sciencedirect.com/science/article/pii/S2352340915001730?via%3Dihub). The data in full is available as four CSV files. However, for this project we will only use data.csv and gps_points.csv.<br /><br />
This data was collected in 2009, from June to early July and "provides field-observed viability status of 1935 trees and surrounding stand conditions." (Oguro et al., 2015) The focus area is in the mountain forests near Tsuruoka City, Yamagata Prefecture, Japan. This particular area was chosen "to cover the variation in land-use types and vegetation, and because of their accessibility." (Oguro et al., 2015) Michio Oguro and her team collected the source data in order to track Japanese Oak Wilt, a disease that causes mass mortality among _fagaceous_ trees in Japan.<br /><br />
The accuracy of the data is sound, as all trees and vegetation were identified and documented by experienced researchers. GPS coordinates were gathered using dedicated GPS devices. The data is highly usable as-is, requiring minimal preparation for our use cases.<br /><br />
Before describing the data in detail, it is helpful to understand a few terms in greater depth.


### **Terms**
_**R. quercivora**_, *Raffaelea quercivora* - a species of fungus known for causing Japanese Oak Wilt<br /><br />
**Japanese Oak Wilt** - a vector-born fungal disease that affects oak trees. It is transmitted by _P. quercivorus_ and has caused excessive death in oak trees in Japan since the 1980s.<br /><br />
_**P. quercivorus**_, _Platypus quercivorus Common Name: Flying Ambrosia Beetle_ - a species of weevil and known vector of _R. quercivora_. _Platypus quercivorus_ carries _R. quercivora_ and boars into host oaks, spreading the fungus. The fungal growth in the host oak becomes Japanese Oak Wilt, killing the host. <br /><br />
_**Q. serrata**_ - Quercus serrata, _Common Name:  Jolcham Oak_ - A species of oak native to China, Taiwan, Japan, and Korea. A common host for _R. quercivora_.<br />
_**C. japonica**_ - _Cryptomeria japonica Common Name: Japanese Cyprus_ , a conifer of the _cyprus_ family<br />
_**F. crenata**_ -  _Fagus crenata Common Name: Japanese Beech_, a tree of the _fagaceae_ family<br />
_**Q. crispula**_ - _Quercus crispula_, a species of oak known in Japan as Mizunara. It is widespread in Northeast Asia. A common host for _R. quercivora_.<br />
<br />
**Breast Height** - a point on the tree trunk at 4.5 feet or 1.37 meters high.<br />
**Basal Area** - the cross-sectional area of a tree at breast height.

### Objective 1 Data Prep
*Map Points*<br /><br />
If our only goal were to map the individual points geographically, we could use the point location data included in gps_points.csv as-is. However, we would like to include additional, useful information, particularly the percentage of trees at each point that have died. We will include this detail not only in the tool-tip for each point but also as a color, using a blue-red diverging gradient as a quick visual indication of the number of trees found dead at each point.<br /><br />
Below, we group `data_df` by point, and generate two columns, the sum (count) of dead trees at each point, and the count of all trees at each point. We rename these new columns for clarity and adjust the indexing.<br /><br />
**Reminder:** While there are other trees documented at each point we are only concerned with those vulnerable to Japanese Oak Wilt, _Q. serrata_ and _Q. crispula_.


```python
# data_df groupby, rename columns, drop multiindex, reset
dead_pt_df = data_df.groupby('point').agg({'dead': ['sum', 'count']}) \
                                     .rename(columns={"sum": "num_dead",
                                                      "count": "num_trees"}) \
                                     .droplevel(0, axis=1) \
                                     .reset_index()
```

Now that we have a dataframe grouped by point with the number of dead trees, as well as total trees at each point, we can calculate the percentage of trees that are dead at each observed point. We merge the resulting data into `gps_df`.

```python
# caclulate % of dead trees per point in new column
dead_pt_df['percent_dead'] = dead_pt_df['num_dead'] / dead_pt_df['num_trees']

# check counts are as expected
dead_pt_df.info()
```

```python
dead_pt_df.head()
```

```python
# join the calculated percentages to gps_df
gps_df = gps_df.merge(dead_pt_df, left_on='point_name', 
                      right_on='point', how='left') \
               .drop(['point'], axis=1)
```

A quick check of `gps_df.info()` after our merge reveals some null values where our counts and percent calculations should be—12 rows total. A quick spot-check of our original `data_df` shows that there are some points logged in gps_df that are not present with tree data in `data_df`. We remove these values before moving forward.

### Objective 2 Data Prep
*Visualize Species Mortality*

We will use a pie chart to view our mortality rates by species relative to the total number of trees observed. To do so, we will prepare a dataframe that represents the count of each species by their "dead" label. Matplotlib can then calculate the percentages for us, as part of the pie chart plotting process.

```python
# create new df with dead and alive counts for each species of tree
dead_count_df = data_df.groupby(['species', 'dead']) \
                        .agg({'dead': 'count'}) \
                        .rename(columns={"dead": "count"})

dead_count_df.reset_index(inplace=True)
```

```python
dead_count_df
```

We will also look into mortality rates as a percentage relative to each species, rather than to all observed trees as a whole. This will give us a better understanding of the impact Japanese Oak Wilt has on each individual species, rather than the entire forest.<br/><br/>
We will prepare a dataframe grouped by species and calculate the percentage of each that are dead and alive. We will then transform the dataframe so that we can pass columns directly to our stacked bar chart.


```python
dead_rate_df = dead_count_df

# calc percent by species and label
dead_rate_df['rel_pct'] = dead_rate_df.groupby(['species'])['count'] \
                                           .apply(lambda x:100 * x / float(x.sum()))
```

```python
dead_rate_df
```

```python
# transform 
dead_rate_df = pd.pivot_table(dead_rate_df,index=['species'], 
                             columns=dead_rate_df.groupby(['species']).cumcount())

# format column names
dead_rate_df.columns = dead_rate_df.columns.map('{0[0]}_{0[1]}'.format)

# drop extraneous columns
dead_rate_df.drop(['dead_0', 'dead_1'], axis = 1, inplace=True)

dead_rate_df.reset_index(inplace=True)
```

```python
dead_rate_df
```

### Objective 3 Data Prep
*Correlation Analysis*<br /><br />
In order to move forward with correlation analysis, we must prepare our data in two particular ways:
 
1.  Scaling
2.  Balancing
 
Currently, a number of our features are set according to different numeric scales. As part of our exploration, we will generate a correlation matrix reliant on a formula that may develop a bias if the scales of our feature values vary greatly. We can avoid the possibility of bias by adjusting all of our values to fit the same scale. In our case, we will use a MinMax scaler and adjust all of our values to fit a 0-1 scale.<ba/><ba/>
It is important to note that when using MinMax scaling, the lowest values become 0.

```python
data_scale_df = data_df.copy()

scaler = MinMaxScaler()

# fit and transform the data
data_scale_df.iloc[:, 3:28] = scaler.fit_transform(data_scale_df.iloc[:, 3:28])
```

```python
data_scale_df
```

Next, we must consider whether or not our data is balanced. We are plotting correlations to our "dead" or "alive" labels. We have previously found that our data is moderately skewed in favor of living trees. Leaving our data imbalanced will likely produce misleading visualizations. We can balance our data by undersampling from our overrepresented label.

```python
# take all 617 observations of dead trees
dead_select = data_scale_df[data_scale_df['dead'] == 1]

# randomly select 617 samples from our living trees
live_select = data_scale_df[data_scale_df['dead'] == 0].sample(n = 617)

# concat into new dataframe
data_scale_bal_df = pd.concat([dead_select, live_select])
```

```python
data_scale_bal_df
```

```python
# check the balance of our classifications
alive_count = data_scale_bal_df['dead'].value_counts()[0]
dead_count = data_scale_bal_df['dead'].value_counts()[1]

alive_percent = round((alive_count / len(data_scale_bal_df)) * 100, 2)
dead_percent = round((dead_count / len(data_scale_bal_df)) * 100, 2)

print(f'# Alive: {alive_count}, {alive_percent}% of dataset')
print(f'# Dead: {dead_count}, {dead_percent}% of dataset')
```

## Objective 3: Correlation Analysis
### Explanation of Results
Our correlation analysis begins with a generalized comparison of each feature to each other using Pearson's Correlation Coefficient. The correlation is measured on a scale of 1 (positive relationship) to -1 (negative relationship) with values at 0 representing no correlation.Our matrix is quite large, but we can open it in a new tab for a clearer view. Before plotting our matrix, we remove <font color='#A30000'>ba.serrata</font> and <font color='#A30000'>ba.crispula</font>. As these trees are the primary target of this study, we expect a strong correlation in the total basal area of these trees and mortality.<br/><br/>
Interestingly, our matrix shows moderately strong positive relationships between features such as <font color='#A30000'>altitude</font> and <font color='#A30000'>fagus.1000</font>—the proportion of beech forests in the circle of 1000 m radius around the tree—indicating that fagus forests may be more prevalent at a certain altitude range. There is also an inverse correlation between <font color='#A30000'>broad.1000</font>—the proportion of broad-leaved forests in the circle of 1000 m radius around the tree--and <font color='#A30000'>fagus.1000</font>, indicating that the two often may not grow together, at least not in the conditions of this forest.<br/><br/>
However, we are interested in the relationships between our features and our dead/alive labels.<br/><br/>


### Visualisation: Corelation Matrix

```python
# drop unnecessary columns
data_corr_df = data_scale_bal_df.drop(['point', 
                                    'species',
                                    'n.serrata', 
                                    'n.crispula', 
                                    'ba.serrata', 
                                    'ba.crispula'], axis=1)
sns.set()

f, ax = plt.subplots(figsize=(30,25))

# view entire dataframe
subset_corr = data_corr_df.corr()

sns.heatmap(subset_corr,
            annot=True, 
            fmt='.2f', 
            cmap='BrBG', 
            annot_kws={'size':8}, 
            vmin=-1, 
            vmax=1)

ax.set_title('Subset Correlation Matrix)', fontsize=14)

plt.show()

subset_corr
```

To gain a better understanding of hour each feature relates to our target, we take a look at the relationships between our dichotomus variable, <font color='#A30000'>dead</font> and each of our continuous variables using a point-biserial correlation coefficient. Before plotting our matrix, we remove <font color='#A30000'>ba.serrata</font> and <font color='#A30000'>ba.crispula</font>. We also adjust our dataset so that we only see relationships to our <font color='#A30000'>dead</font> variable.<br/><br/>
Viewing our row <font color='#A30000'>dead</font>, we find that there are no strong or even moderate correlations to explore.<br/><br/>
Let's adjust our standards and explore correlations at a lower threshold. We find <font color='#A30000'>ba</font>, <font color='#A30000'>fagus.1000</font>, <font color='#A30000'>grass.1000</font>, and <font color='#A30000'>altitude</font> may warrant a closer look.

```python
from scipy.stats import pointbiserialr

# get continuous and dichotomous data
dead_class = data_corr_df['dead']


new_corr = {'dead': 1}

for col in data_corr_df.columns:
  to_corr = data_corr_df[col]

  # pbc of first question
  pbc = pointbiserialr(dead_class, to_corr)
  new_corr[col] = pbc[0]

new_corr_df = pd.DataFrame(new_corr, index=['dead'])
new_corr_df
```

```python
sns.set()

f, ax = plt.subplots(figsize=(30, 5))

sns.heatmap(new_corr_df,
            annot=True, 
            fmt='.2f', 
            cmap='BrBG', 
            annot_kws={'size':8}, 
            vmin=-1, 
            vmax=1)

ax.set_title('Subset Correlation Matrix)', fontsize=14)

plt.show()
```

### Visualisation: Scatter/Regression Plots
Focusing on the features with the highest correlation to <font color='#A30000'>dead</font> above, we can more apparently see the relationships using a scatter plot with a line of regression.
* **ba:** Dead trees seem to have a larger basal area.

* **grass.1000:** As the proportion of grassland within 1000m of our points increases, the number of dead trees seems to decrease.

If there was no relationship between two features whatsoever, we would expect a flat line of regression. ### Visualisation: KDE Plot by Feature
The following KDE plots indicate that:

In the final visualization, we view <font color='#A30000'>altitude</font> by species for a more in-depth understanding.

# Conclusion
 
### Achievements
Unsurprisingly, we were able to show that Japanese Oak Wilt is likely to affect other trees in the vicinity of host trees. More interestingly, we also demonstrated that _Q.crispula_ trees are much more vulnerable to Japanese Oak Wilt than _Q. serrata_ trees. Although our correlation analysis did not indicate any environmental relationships that humans could leverage to intervene in the spread of Japanese Oak Wilt, we did find a possible explanation as to why _Q. serrata_ is less likely to die from Japanese Oak Wilt—perhaps the altitude at which _Q. serrata_ trees tend to grow also correlates negatively to the altitude at which our vector, the Flying Ambrosia Beetle, tends to burrow. Further data and exploration would be necessary.
 
### Limitations
 
This project had a number of limitations. Foremost is the size and scope of our dataset in terms of the number of observations collected and the area in which the research took place. A larger, more varied collection area may mitigate potential discrepancies caused by nearby cities, footpaths, and people. Second is the lack of associated data on the vector itself. As mentioned above, information such as altitude at which Flying Ambrosia Beetles burrow may be significant. Unfortunately, we only know burrowing details in the form of whether or not a tree had died to Japanese Oak Wilt. Additional details that would benefit this dataset may include trees that show signs of Flying Ambrosia Beetles burrows in addition to their viability. That would allow us to better understand trees that had been or could potentially have been infected, but survived—assuming that is even possible.
 
### Future Work
For the benefit of future research, we would advocate for obtaining similar data from additional points across the forests of Japan, perhaps at a larger range of altitudes and additionally with climate data. We also would like to obtain data not just on the viability of each observed tree, but whether or not signs that the tree had been visited by our vector are present. Finally, additional demographic data on _Q.serrata_ and _Q. crispula_ may help us pinpoint why they die to Japanese Oak Wilt at such different rates. The culmination of this more thorough data collection may lead to an adjustment of priorities and perhaps additional questions that could lead to a viable human intervention.
